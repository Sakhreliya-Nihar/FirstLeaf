```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
image = cv2.imread("tnt.png")

if image is None:
  print("Error: Unable to load the image.")
else:
  rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```
gray_image = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```
pixels = rgb.reshape((-1, 3))
pixels = np.float32(pixels)
K = 8
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
ret, label, center = cv2.kmeans(pixels, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
centers = np.uint8(center)
data_rgb = centers[label.flatten()]
data_rgb = data_rgb.reshape(rgb.shape)
plt.imshow(data_rgb)
plt.title('RGB to Indexed Image ')
plt.axis('off')
plt.show()
```

```
gray_pixels_flattened = gray_image.reshape((-1, 1))
gray_pixels_float = np.float32(gray_pixels_flattened)
K = 4
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
ret, label, center_gray = cv2.kmeans(gray_pixels_float, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
centers_gray = np.uint8(center_gray)
indexed_gray = centers_gray[label.flatten()]
indexed_gray = indexed_gray.reshape(gray_image.shape)
plt.imshow(indexed_gray, cmap='gray')
plt.title("gray to index")
plt.axis('off')
plt.show()
```
